```python
resANS = []
for end in ends:
    for inter in range(5,55,5):
        event_file = end+str(inter)+"/"
        tmp = event_file.split('/')
        agent_type = tmp[7]
        agent_tech = tmp[8][:2]
        n_maq = tmp[9][:1]
        dim = tmp[10]

        event_acc = EventAccumulator(event_file)

        event_acc.Reload()

        prod_per = []
        for e in event_acc.Scalars('Production/per'):
            prod_per.append(e.value)

        max_prod = np.max(prod_per)
        mean_prod = np.mean(prod_per)
        var_prod = np.var(prod_per)

        act_0 = []
        act_1 = []
        for e in event_acc.Scalars('vars/Action_0'):
            act_0.append(e.value)
        for e in event_acc.Scalars('vars/Action_1'):
            act_1.append(e.value)
        var_act_0 = np.var(act_0)
        var_act_1 = np.var(act_1)

        rw = []
        for e in event_acc.Scalars('Reward/Step reward'):
            rw.append(e.value)

        max_rw = np.max(rw)
        mean_rw = np.mean(rw)
        var_rw = np.var(rw)

        event_file+'Production_M2_wp'

        event_wp1 = EventAccumulator(event_file+'Production_M1_wp')
        event_wp1.Reload()
        event_wp2 = EventAccumulator(event_file+'Production_M2_wp')
        event_wp2.Reload()

        wp_1 = []
        wp_2 = []
        for e in event_wp1.Scalars('Production/M1'):
            wp_1.append(e.value)
        for e in event_wp2.Scalars('Production/M2'):
            wp_2.append(e.value)

        max_wp_1 = np.max(wp_1)
        mean_wp_1 = np.mean(wp_1)
        var_wp_1 = np.var(wp_1)

        max_wp_2 = np.max(wp_2)
        mean_wp_2 = np.mean(wp_2)
        var_wp_2 = np.var(wp_2)

        resRun = [str(inter),agent_type,agent_tech,n_maq,dim,\
                  max_prod,mean_prod,var_prod,\
                  max_rw,mean_rw,var_rw,\
                  max_wp_1,mean_wp_1,var_wp_1,\
                  max_wp_2,mean_wp_2,var_wp_2,\
                  var_act_0,var_act_1]

        resANS.append(resRun)

        fileRes = "Tab_Res_Ans.csv"
        with open(fileRes, 'a') as csvfile:
            csvwriter = csv.writer(csvfile)
            csvwriter.writerow(resRun)
            csvfile.close()
```

```python
df = pd.DataFrame(resANS)
df.columns =['Teste', 'Agente', 'Tipo','N maq', 'Tam Net',\
             'Max Prod', 'Media Prod','Var Prod',\
             'Max Rec',  'Media Rec', 'Var Rec',\
             'Max Wp1',  'Media Wp1', 'Var Wp1',\
             'Max Wp2',  'Media Wp2', 'Var Wp2',\
             'Var act1', 'Var ac2']
```
